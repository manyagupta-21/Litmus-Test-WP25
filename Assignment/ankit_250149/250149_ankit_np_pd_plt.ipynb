```python
# Task 2.1: Create a DataFrame from the data of students above
```

```python
d = {"Roll No." : [2301, 2310, 2302, 2304, 2306, 2305, 2307, 2312, 2315, 2316, 2320, 2308, 2309, 2314, 2319], "Name" : ["Alice", "Smith", "Bob", "Charlie", "Eric", "Diana", "John", "Joey", "Bean", "Charlie", "Jack", "Jill", "George", "Tabitha", "Phoebe"], "Age":[24, None, 22, 23, None, 24, None, 30, 26, 23, 22, 21, 24, 31, 24], "Marks":[85, None, 78, 92, 95, 88, None, 64, 72, 92, 83, 69, 89, 100, 91]}
df = pd.DataFrame(d).set_index("Roll No.")
df
```

```python
# (BONUS, but really fun try to find it do it how :) ) Task 2.6: Make a function to assign grade
# (i) A if marks > 85
# (ii) B if marks > 75
# (iii) C if marks > 65
```

```python
grades = []
for i in range(0,len(df)):
  if (df.iloc[(i,2)]> 85) :
    grades.append("A")
  elif (df.iloc[(i,2)]> 75) :
    grades.append("B")
  elif (df.iloc[(i,2)]> 65) :
    grades.append("C")
  else:
    grades.append("NaN")

df["Grades"] = grades
df
```

```python
# Task 2.7: Rename Marks column to Score
```

```python
df.rename(columns={"Marks":"Score"}, inplace = True )
df
```

```python
# Task 2.8: Replace NaN in Age with mode and NaN in Marks with mean
```

```python
for i in range(0,len(df)):
  if (str(df.iloc[(i,1)]) == "nan"):
    df.iloc[(i,1)] = df["Age"].mode().iat[0]
  if (str(df.iloc[(i,2)]) == "nan"):
    df.iloc[(i,2)] = df["Score"].mean()
df
```

# ***Part 3: Matplotlib***

```python
import matplotlib.pyplot as plt
```

```python
# Task 3.1: Plot a Pie Chart depicting the percentag of grades that the students have got.
```

```python
grade_counts = df["Grades"].value_counts()
grade_percent = grade_counts *100 / grade_counts.sum()

plt.pie(grade_percent, labels =grade_percent.index, startangle=90)
plt.show()
```

```python
# Task 3.2: Plot a Line Graph of Roll No. vs Marks
```
